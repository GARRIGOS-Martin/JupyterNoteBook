```python
# Combien d’arbres ont été plantés année par année (ex : 1987 : 771, 1988 : 266, etc…) ?

# La fonction set est très intéressante surtout pour créer un dictionnaire !

# Méthode 1 : on affiche tout ça dans un dictionnaire

def nombre_arbre_plantes_par_annee(liste) :
    arbres_plantes = {}
    for annee in set(liste) :
        arbres_plantes[annee] = liste.count(annee)
    return arbres_plantes
        
apa = nombre_arbre_plantes_par_annee(annee_plantation_plein)
print(apa)

# Méthode 2 : on affiche tout ça avec des print

for annee in set(annee_plantation_plein) :
    arbres_plantes = annee_plantation_plein.count(annee)
    print (annee, " : ", arbres_plantes)

```
